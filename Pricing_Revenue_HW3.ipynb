```python
import pandas as pd
import numpy as np
from scipy.optimize import minimize

# load data from InSample Q1.xlsx file
data = pd.read_excel('InSample_Q1.xlsx').values

# define log-likelihood function
def log_likelihood(beta, data):
    beta0, beta1 = beta
    loglik = 0
    for i in range(len(data)):
        pi, yi = data[i]
        gamma = np.exp(beta0 - beta1*pi)/(1 + np.exp(beta0 - beta1*pi))
        loglik += yi*np.log(gamma) + (1 - yi)*np.log(1 - gamma)
    return -loglik

# maximize log-likelihood function to find beta0* and beta1*
initial_guess = [0, 0]
result = minimize(log_likelihood, initial_guess, args=(data,))
beta0_star, beta1_star = result.x

# print estimated values of beta0* and beta1*
print('beta0*:', beta0_star)
print('beta1*:', beta1_star)
```

```python
import scipy.optimize as sp
# Load the data
data = pd.read_excel("InSample_Q1.xlsx", names=["p_i", "y_i"])

# Extract the price and sold/not sold data
p = data['p_i']
y = data['y_i']
# Define the logit demand function
def logit_demand(price, beta0, beta1):
    u = beta0 - beta1 * price
    gamma = np.exp(u) / (1 + np.exp(u))
    return gamma

# Define the revenue function
def revenue(beta0, beta1):
    return -p * logit_demand(p, beta0, beta1) * len(y)

# Find the optimal price numerically
optimal_price = sp.minimize_scalar(revenue(beta0_star, beta1_star), bounds=(0, None)).x

# Read in the out-of-sample data
out_of_sample_data = pd.read_excel('OutOfSample_Q1.xlsx')

# Compute the revenue for the optimal price
expected_revenue = revenue(optimal_price, beta0_star, beta1_star)
revenue_out_of_sample = out_of_sample_data['WTP'].apply(lambda p: revenue(p, beta0_star, beta1_star)).sum()


print(f"Optimal price: ${optimal_price:.2f}")
print(f"Expected revenue: ${expected_revenue:.2f}")
print(f"Revenue in out-of-sample data: ${revenue_out_of_sample:.2f}")
```
